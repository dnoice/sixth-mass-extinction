## üéØ Objective
To visualize the **cumulative number of extinctions** over time (1500-2015). This notebook aims to demonstrate the acceleration of the extinction rate (the "Inflection Point") associated with the industrial era.

## üî¨ Methodology
We reconstruct the time-series using two pivotal data points from Ceballos et al. (2015):
1.  Total Extinctions since 1500.
2.  Total Extinctions since 1900.

We compare this observed curve against the expected linear accumulation of the Background Rate (2 E/MSY).

---

```python
# üõ†Ô∏è Setup & Imports
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path

plt.style.use('seaborn-v0_8-whitegrid')
DATA_PATH = Path('data')
OUTPUT_PATH = Path('data')

# Ensure local data exists
if not (DATA_PATH / 'ceballos_2015.csv').exists():
    raise FileNotFoundError("Required data not found.")
```

```python
# üì• Data Construction

# Reconstructing the time series from Ceballos (2015) summary stats
# Total Vertebrates (High Scenario): 617 extinctions (1500-2015)
# Extinctions post-1900: ~477 (Derived from 53x elevation rate)

points = [
    {'Year': 1500, 'Cumulative_Extinctions': 0, 'Type': 'Observed'},
    {'Year': 1900, 'Cumulative_Extinctions': 140, 'Type': 'Observed'}, # 617 total - 477 post-1900
    {'Year': 2015, 'Cumulative_Extinctions': 617, 'Type': 'Observed'}
]

ts_df = pd.DataFrame(points)
print("Observed Trend (Reconstructed):")
print(ts_df)
```

```python
# üìä Calculating Background Trend
# Background = 2 E/MSY
# Total Vertebrates Assessed ~ 39,223 (Ceballos 2015)

bg_points = []
TOTAL_ASSESSED = 39223
BACKGROUND_RATE = 2.0

for year in range(1500, 2020, 10):
    years_elapsed = year - 1500
    expected = (BACKGROUND_RATE * TOTAL_ASSESSED * years_elapsed) / 1_000_000
    bg_points.append({'Year': year, 'Cumulative_Extinctions': expected, 'Type': 'Background (2 E/MSY)'})

bg_df = pd.DataFrame(bg_points)
combined_df = pd.concat([ts_df, bg_df])
```

```python
# üìà Visualization: The Hockey Stick

plt.figure(figsize=(10, 6))
sns.lineplot(data=bg_df, x='Year', y='Cumulative_Extinctions', color='green', label='Background Rate (2 E/MSY)')
sns.lineplot(data=ts_df, x='Year', y='Cumulative_Extinctions', color='red', marker='o', label='Observed (Vertebrates)')

plt.title('The Hockey Stick: Cumulative Vertebrate Extinctions (High Scenario)')
plt.ylabel('Cumulative Extinctions')
plt.xlabel('Year')
plt.legend()
plt.grid(True, alpha=0.3)

plt.annotate('Industrial Acceleration
(1900)', xy=(1900, 140), xytext=(1700, 300),
             arrowprops=dict(facecolor='black', shrink=0.05))

plt.savefig(OUTPUT_PATH / 'cumulative_hockey_stick.png')
plt.show()
```
